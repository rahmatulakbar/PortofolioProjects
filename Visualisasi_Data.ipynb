```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
df=pd.read_csv('/content/drive/MyDrive/Data/Car_sales.csv')
df.head()
```

```python
#jenis plot
style.use('ggplot')
#ukuran 
plt.figure(figsize=(10,7))
#plot histogram
plt.hist(df.Sales_in_thousands, bins=20)
#nama histogram 
plt.title('histogram')
#menampilkan plot
plt.show()
```

```python
#ukuran gambar
plt.figure(figsize=(15,6))
#jenis plot
style.use('ggplot')
#plot data
plt.plot(df.Price_in_thousands,'r',label='harga mobil',linewidth=2)
#label sumbu x
plt.xlabel('merk')
#label sumbu y
plt.ylabel('harga')
#nama plot
plt.title('plot data')
#menampilkan legend
plt.legend()
#menampilkan plot
plt.show()
```

```python
#ukuran gambar
plt.figure(figsize=(15,6))
#jenis plot
style.use('ggplot')
#plot data pertama
plt.plot(df.Price_in_thousands,'r',label='harga mobil',linewidth=2)
#plot data kedua
plt.plot(df.Sales_in_thousands,'b',label='Penjualan',linewidth=2)
#label sumbu x
plt.xlabel('merk')
#label sumbu y
plt.ylabel('harga')
#nama plot
plt.title('plot data')
#menampilkan legend
plt.legend()
#menampilkan plot
plt.show()
```

```python
#jenis plot dan ukuran 
style.use('ggplot')
plt.figure(figsize=(10,7))
#scatter plot
plt.scatter(df.Engine_size,df.Price_in_thousands,label='harga mobil')
#nama plot
plt.title('Scatter Plot')
#label sumbu x
plt.xlabel('ukuran mesin')
#label sumbu y
plt.ylabel('harga')
#menampilkan legenda
plt.legend()
#menampilkan plot
plt.show()
```

```python
#manyusun data berdasarkan jenis manufactur nya
df1=df.groupby('Manufacturer').sum()
df1.head()
```

```python
plt.figure(figsize=(10,10))
#plot data
plt.pie(df1.Sales_in_thousands,labels=df1.index,autopct='%1.1f%%')
plt.axis('equal')
plt.title('pie chart')
plt.show()
```

```python
plt.figure(figsize=(15,6))
#plot data pertama
plt.subplot(1,2,1)
plt.plot(df.Sales_in_thousands,'b')
plt.title('Harga Mobil')
#plot data kedua
plt.subplot(1,2,2)
plt.plot(df.Horsepower,'g')
plt.title('Jumlah Tenaga Kuda')
plt.show()
```

```python
plt.figure(figsize=(15,6))
#plot data pertama
plt.subplot(2,1,1)
plt.plot(df.Width,'r')
plt.title('Lebar Mobil')
#plot data kedua
plt.subplot(2,1,2)
plt.plot(df.Length,'y')
plt.title('Panjang Mobil')
plt.show()
```
