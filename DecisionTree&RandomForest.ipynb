```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('/content/drive/MyDrive/Data/IBM Attrition Data.csv')
df.head()
```

```python
df.isna().sum()
```

```python
X=df.drop(['Attrition'],axis=1)
y=df[['Attrition']]
```

```python
X=pd.get_dummies(X)
X.head()
```

```python
print(df.columns)
print(X.columns)
```

```python
Lb=LabelEncoder()
y=Lb.fit_transform(y)
```

```python
Xtrain,Xtest, ytrain,ytest=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

classifier=DecisionTreeClassifier()
model=classifier.fit(Xtrain,ytrain)

```

```python
prediksi=model.predict(Xtest)
accuracy_score(ytest,prediksi)
```

```python
from sklearn.ensemble import RandomForestClassifier
Rf=RandomForestClassifier()
model1=Rf.fit(Xtrain,ytrain)
```

```python
prediksi1=model1.predict(Xtest)
accuracy_score(ytest,prediksi1)
```
